```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_data.csv')
```

# Numerical-Numerical

## Scatterplot

```python
sns.scatterplot(x = 'studytime',y='G3',hue = 'sex',data=df)
plt.show()
```

The scatter plot shows the relationship between study time and final grades (G3). Students who spend more time studying tend to achieve higher grades. The performance distribution for male and female students appears relatively similar across different study time levels.

```python
sns.scatterplot(x = 'absences',y='G3',data=df)
plt.show()
```

A higher number of absences may negatively affect student performance, and grades show greater variability.


```python
sns.scatterplot(x = 'G1',y='G3',data=df)
plt.show()
```

```python
sns.scatterplot(x = 'G2',y='G3',data=df)
plt.show()
```

The first period grade (G1) and second period grade (G2) show a strong relationship with the final grade (G3).

# Numerical-Categorical

```python
sns.boxplot(x ='school',y='G3',data=df)
plt.show()
```

Some variation in student performance can be observed between different schools.

```python
sns.boxplot(x ='sex',y='G3',data=df)
plt.show()
```

The distribution of grades for male and female students appears relatively similar.

```python
sns.boxplot(x ='address',y='G3',data=df)
plt.show()
```

There may be slight differences in performance between students from urban and rural areas.

```python
sns.boxplot(x ='famsize',y='G3',data=df)
plt.show()
```

Family size does not show a strong or clear impact on student performance.

```python
sns.boxplot(x ='Pstatus',y='G3',data=df)
plt.show()
```

# Categorical-Categorical

```python
sns.countplot(x='school', hue='sex', data=df)
plt.show()
```

The distribution of male and female students varies across the two schools.

```python
sns.countplot(x='address', hue='famsize', data=df)
plt.show()
```

Family size distribution differs between urban and rural areas.

```python
sns.countplot(x='sex', hue='Pstatus', data=df)
plt.show()
```

The parental status distribution shows variation across genders

## Observations:

Student academic performance appears to be influenced by several factors such as:

--Study time

--Past failures

--Absences

--Previous academic performance (G1 and G2)

Among these variables, G1 and G2 show the strongest relationship with the final grade (G3).
